```python
import requests
```

```python
l = list(range(100,8600,100))
l[0]
```

```python
url_start = "http://amis.co.ke/site/market/{}?product=1&per_page=100.html"
```

```python
# if you get this error do it as below------
# "b'<head><title>Not Acceptable!</title></head><body><h1>Not Acceptable!</h1><p>An appropriate 
# representation of the requested resource could not be found on this server. 
# This error was generated by Mod_Security.</p></body></html>
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
# start = requests.get(url, headers=headers)
```

```python
# start.text
```

```python
for y in l:
    url = url_start.format(y)
    
    data = requests.get(url, headers=headers)
    
    with open("mpr/{}.html".format(y), "w+") as f:
        f.write(data.text)
```

```python
dfs = []
for year in l:
    with open("mpr/{}.html".format(year)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    mpr_table = soup.find_all(id="nav-data")[0]
    mpr_df = pd.read_html(str(mpr_table))[0]
    dfs.append(mpr_df)
```

```python
mprs = pd.concat(dfs)

mprs.head()
```

```python
mprs.to_csv("mprs.csv")
```
